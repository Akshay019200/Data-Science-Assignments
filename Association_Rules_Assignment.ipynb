# **2. Association Rule Mining:**

```python
#pip install apyori
```

```python
# Apriori package accepts only list so we will create an empty list and append all the transaction

from apyori import apriori


trans= []

for i in range(0, 7501):
    trans.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
trans
```

```python
# To get the rules from the below parameters

rules= apriori(transactions= trans,
              min_support=0.003,
              min_confidence= 0.2,
              min_lift= 3,
              min_length= 2, max_length= 2)
rules
```

```python
# Converting the rules into a list and saving it in results

results= list(rules)
results
```

```python
# Checking for the length of results

len(results)
```

```python
# 1st rule

results[0]
```

```python
# Creating empty list

base_item=[]
add_item=[]
support=[]
confidence=[]
lift=[]
```

```python
# To access the required element from the results

results[0][2][0][0]   #Base item
results[0][2][0][1]   #Add item
results[0][1]         #Support
results[0][2][0][2]   #Confidence
results[0][2][0][3]   #Lift
```

```python
# Using for loop to add all the required elements to the above empty list

for i in range(0,9):
    base_item.append(results[i][2][0][0])
    add_item.append(results[i][2][0][1])
    support.append(results[i][1])
    confidence.append(results[i][2][0][2])
    lift.append(results[i][2][0][3])
```

```python
# Creating a dataframe of all the elements

pd.DataFrame(base_item)
pd.DataFrame(add_item)
pd.DataFrame(support)
pd.DataFrame(confidence)
pd.DataFrame(lift)
```

```python
# Concatenating all the above dataframe and saving it under new_data

new_data=pd.concat([pd.DataFrame(base_item),pd.DataFrame(add_item),pd.DataFrame(support),pd.DataFrame(confidence),pd.DataFrame(lift)], axis=1)
```

```python
new_data.columns= ['Base item', 'Add item', 'Support', 'Confidence', 'Lift']
new_data
```

1. What is lift and why is it important in Association rules?

> Lift measures how much more likely two items are to be bought together compared to how often they are bought individually.

It's important because it helps identify meaningful associations between items. A high lift value indicates a strong relationship between items, which can be useful for businesses to understand customer behavior and optimize their strategies, like product placement or marketing.

2. What is support and Confidence. How do you calculate them?

> Support: It refers to the frequency or occurence of a particular set of items in a dataset. It indicates how often a specific combination of items appears together in the data. Higher support values suggest stronger relationships or associations between the items, making them more significant for generating association rules.

Support is calculated by dividing the number of transactions containing the item set by the total number of transactions.

Confidence: It is the measure that shows how often two things (like items in a dataset) tend to occur together. It helps us understand the likelihood of one thing happening when the other occurs. Higher confidence means a stronger connection between two items.

Confidence is calculated by dividing the number of transactions containing both items in the rule by the number of transactions containing the first item.

confidence(A/B)= (A intersection B)/B (Conditional probability)

3.	What are some limitations or challenges of Association rules mining?

> One challenge is the presence of spurious or random associations, especially in large datasets. Not all associations identified are meaningful or actionable.

Association rules don't consider the order of item occurrence in transactions, which can limit their applicability in certain scenarios.

They may not capture complex relationships or dependencies between items, particularly when dealing with a large number of items.

Association rules mining requires a significant amount of data to generate reliable results, and the interpretation of rules often requires domain knowledge to be meaningful.
